```python
#Importamos librerías

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import sklearn
```

```python
#leemos el dataset
dataset = pd.read_csv('C:/ClusterAI/Clustering/Mall_Customers.csv')
# Nos quedamos con las variables importantes para hacer el análisis.
X = dataset.iloc[:,[3,4]].values
```

* **Utilizaremos el método del codo para encontrar el número óptimo de cluster**
* **WCSS es la suma de la distancia al cuadrado entre cada punto y su centroide en un cluster** 

```python
#importamos Kmeans
from sklearn.cluster import KMeans

#Probaremos de 1 a 10 cluster para ver cual funciona mejor

wcss = [] 
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
```

### Método del codo

```python
plt.plot(range(1,11), wcss)
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.title('Gráfico del codo')
plt.show()
```

Cual creen que sería el número óptimo de clusters? Por que?

```python
## Entrenamos y predecimos con k = 5
kmeans = KMeans(n_clusters=5, init = "k-means++", random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

Graficamos los distintos clusters y sus centroides para ver como quedaron determinados

```python
#  Ploteamos los puntos y los distintos clusters
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1], s=60, c='red', label = 'Cluster 1')  
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1], s=60, c='blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1], s=60, c='green', label = 'Cluster 3')
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1], s=60, c='violet', label = 'Cluster 4')
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1], s=60, c='yellow', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'black', label = 'Centroides')
plt.legend()
plt.show()
```
